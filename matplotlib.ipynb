```python
import matplotlib.pyplot as plt
salaries=[15,18,20,28,90,76,89,]
plt.hist(salaries,bins=6,color='purple',edgecolor='black',alpha=0.7)
plt.xlabel("salary (in thousands)")
plt.ylabel("salaries distribution of employees")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[0,2,4,6,8]
y=[0,10,20,30,40]
plt.plot(x,y,color='blue',marker='o',linestyle='-')
plt.title("Simple line plot")
plt.xlabel("x-axis")
plt.ylabel("x-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
branch=['cse','it','eee','ece']
pass_percentage=[92,88,81,76]
colors=['red','blue','green','orange']
plt.bar(branch,pass_percentage,color=colors)
plt.title("second year")
plt.xlabel("branch")
plt.ylabel("pass percentage")
plt.show()
```

```python
import matplotlib.pyplot as plt
sport=['crickets','footbal','badmanton','volleyball']
student=[120,80,60,40]
plt.pie(student,labels=sport,autopct='%1.1f%%',startangle=90)
plt.title("student sports")
plt.show()
```

```python
import matplotlib.pyplot as plt
attendance=[60,66,45,80,93]
marks=[55,76,90,33,54]
plt.scatter(attendance,marks,s=120,c='blue',marker='*',alpha=0.8,label='student data')
plt.xlabel("attendance percentage")
plt.ylabel("marks obtained")
plt.title("scatter plot:attendance")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
marks=[55,76,90,33,54]
plt.boxplot(marks,patch_artist=True)
plt.ylabel("marks")
plt.title("box plot of student marks")
plt.show()
```

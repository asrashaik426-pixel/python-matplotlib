```python
import seaborn as sns

import matplotlib.pyplot as plt

import pandas as pd

# Create sample data

data = pd.DataFrame({

"Attendance": [60, 65, 70, 75, 80, 85, 90, 95],

"Marks": [50, 55, 58, 65, 70, 75, 82, 90]

})

# Create lmplot

sns.lmplot(x="Attendance", y="Marks", data=data)

plt.title("Relationship between Attendance and Marks")

plt.show()
```

```python
import seaborn as sns

import matplotlib.pyplot as plt

# Load the dataset

data = sns.load_dataset("iris")

# Create the impplot

sns.lmplot(x="sepal_length",

y="petal_length", data=data)

# Add labels and title

plt.xlabel("sepal_length")

plt.ylabel("petal_length")

plt.title("Implicitly Connected Scatter Plot of petal lenth by sepal length")

plt.show()
```

```python
import seaborn as sns

import matplotlib.pyplot as plt

# Load dataset

data = sns.load_dataset("iris")

# Create distplot

sns.distplot(data["sepal_length"])

plt.title("Distribution Plot Example")

plt.show()
```

```python
import seaborn as sns

import matplotlib.pyplot as plt

# Load the dataset

iris = sns.load_dataset("iris")

# Create the distplot

sns.distplot(iris[iris["species"] == "setosa"]["sepal_width"], label="Setosa", hist=False)

sns.distplot(iris[iris["species"] == "versicolor"]["sepal_width"], label="Versicolor", hist=False)

sns.distplot(iris[iris["species"] == "virginica"]["sepal_width"], label="Virginica", hist=False)

# Add labels and title

plt.xlabel("Sepal Width")

plt.ylabel("Density")

plt.title("Distribution of Sepal Width by Iris Species")

# Show the plot

plt.legend()

plt.show()
```

```python
import seaborn as sns

import matplotlib.pyplot as plt

import pandas as pd

# Create dataset

data = {

"Department": ["CSE", "ECE", "MECH", "CSE", "ECE", "MECH", "CSE", "ECE", "MECH"],

"Marks": [80, 75, 70, 85, 78, 72, 90, 82, 74]

}

df = pd.DataFrame(data)

# Create cat plot

sns.catplot(x="Department", y="Marks", data=df, kind="strip")

# Add labels and title

plt.xlabel("Department")

plt.ylabel("Marks")

plt.title("Cat Plot of Student Marks by Department")

# Display plot

plt.show()
```

```python
import seaborn as sns

import matplotlib.pyplot as plt

# Load the dataset

data = sns.load_dataset("iris")

# Create the cat plot

sns.catplot(x="species", y="sepal_width", data=data, kind="strip", height=4, aspect=1.5)

# Add labels and title

plt.xlabel("Iris Species")

plt.ylabel("Sepal Width")

plt.title("Strip Plot of Sepal Width by Iris Species")

# Show the plot

plt.show()
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
a=sns.load_dataset("iris")
d=sns.swarmplot(x="species",y="sepal_width",data=a,hue='species')
d.set(xlabel="iris",ylabel="sepal width")
plt.title("swarm plot of species in iris")
plt.show()
```
